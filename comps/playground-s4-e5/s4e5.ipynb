Evaluated using R2 score



```python
import pandas as pd
import numpy as np
```

```python
TRAIN = pd.read_csv("./train.csv").drop(columns=["id"])
TEST = pd.read_csv("./test.csv").drop(columns=["id"])
SUBMISSION = pd.read_csv("./sample_submission.csv")

floodTrain = TRAIN[["FloodProbability"]]
TRAIN = TRAIN.drop(columns=["FloodProbability"])

TRAIN.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

ss = StandardScaler()
n_train = ss.fit_transform(TRAIN)
n_test = ss.transform(TEST)

n_train = pd.DataFrame(columns=TRAIN.columns, data=n_train)
n_test = pd.DataFrame(columns=TEST.columns, data=n_test)

x_train, x_test, y_train, y_test = train_test_split(n_train, floodTrain, train_size=.8)
```

```python
from lightgbm import LGBMRegressor


model = LGBMRegressor(n_estimators=1000, eval_metric="r2")

model.fit(x_train, y_train, eval_metric = "r2", eval_set = [(x_test, y_test)])

preds = model.predict(x_test)
print(r2_score(y_test, preds))

y_pred = model.predict(n_test)
```

```python
SUBMISSION["FloodProbability"] = y_pred

SUBMISSION.to_csv("./submissions/May1Sub1.csv", index=False)
```
